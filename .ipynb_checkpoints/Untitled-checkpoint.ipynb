```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
from colorspacious import cspace_converter
from collections import OrderedDict
```

```python
info_doc=["Category","Type of User","Technology General","Specific Application","Wireless Technology","Mode of Communication","KPI","Requirement","Type of Road"]
number1=[57,3,9,73,75,73,64,43,18]
number=[i*100/108 for i in number1]
```

```python
df = pd.DataFrame({'info': info_doc,"number":number})
df=df.sort_values(by=['number'])
```

```python
name = df['info'].head(12)
price = df['number'].head(12)
 
```

```python

cmaps = OrderedDict()
```

```python
cmaps['Sequential'] = [
            'Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
            'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
            'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn']
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))
 
# Horizontal Bar Plot
bar=ax.barh(name, price)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
plt.xlabel("Percentage") 
# Add Plot Title
ax.set_title('Number of Documents',
             loc ='center', )

color=['yellow','lightsteelblue','indigo','black','lightgrey','darkolivegreen','goldenrod','orangered','skyblue','royalblue']
for i in range(0,len(bar)):
    bar[i].set_color(color[i])
   


# Show Plot

plt.savefig('document_info.pdf')  
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
N=len(df1)
boys = df1['Conference']
girls = df1["Journal"]
omline=df1["Online resource"]
report=df1["Report"]
ind = np.arange(N)  
width = 0.35 
 
fig, ax = plt.subplots(figsize =(16, 9))
p1 = plt.bar(ind, boys, width,)
p2 = plt.bar(ind, girls, width,
             bottom = boys,)
p3 = plt.bar(ind, omline, width,)
p4 = plt.bar(ind, report, width,
             bottom = boys,)
 
plt.ylabel('Number of documents')
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
plt.xticks(ind, df1["Années"])
plt.yticks(np.arange(0, 20, 1))
plt.legend((p1[0], p2[0],p3[0], p4[0]), ('Conference', 'Journal',"Online resource","Report"))
 
plt.savefig('document_type.pdf')  
```
