```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
from colorspacious import cspace_converter
from collections import OrderedDict
```

```python
info_doc=["Category","Type of User","Technology General","Specific Application","Wireless Technology","Mode of Communication","KPI","Requirement","Type of Road"]
number1=[57,3,9,73,75,73,64,43,18]
number=[i*100/108 for i in number1]
```

```python
df = pd.DataFrame({'info': info_doc,"number":number})
df=df.sort_values(by=['number'])
```

```python
name = df['info'].head(12)
price = df['number'].head(12)
 
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(15, 10))
 
# Horizontal Bar Plot
bar=ax.barh(name, price,height=0.8)

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 1)
ax.yaxis.set_tick_params(pad = 1)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 'large', fontweight ='bold',
             color ='grey')
plt.xlabel("Percentage",fontsize='large') 
# Add Plot Title
#ax.set_title('Number of Documents',
    #         loc ='center', )

color=['yellow','lightsteelblue','indigo','black','lightgrey','darkolivegreen','goldenrod','orangered','skyblue','royalblue']
for i in range(0,len(bar)):
    bar[i].set_color(color[i])
 

# Show Plot

plt.savefig('document_info.pdf')  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6.2, 10), subplot_kw=dict(aspect="equal"))

recipe = Nombre

data = Nombre
ingredients = Name


def func(pct, allvals):
    absolute = int(round(pct/100.*np.sum(allvals)))
    return "{:.1f}%".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          title="Content",
          loc="center left",
          bbox_to_anchor=(0.27, -0.1, 0, 0))

plt.setp(autotexts, size=15, weight="bold")

plt.savefig('pie_chart.pdf')  

plt.show()
```
